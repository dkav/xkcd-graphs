## Plot using python and matplotlib

```python
%pylab inline
```

```python
plt.xkcd()
fig, ax = plt.subplots()

x = np.linspace(0, 10, 100)

ax.spines['bottom'].set_position('center')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlim(0, 10)
ax.set_ylim(-1.5, 1.5)

plt.plot(x, sin(x)*-1,'b', lw=1, label='Blue')
plt.plot(x, sin(x),'r', lw=1, label='Red')

plt.xlabel('X')
plt.ylabel("Y")
plt.title('matplotlib xkcd')
plt.legend(loc='upper right', bbox_to_anchor = (1.1, 0.1))
```

## Plot using R and ggplot2

```python
%reload_ext rpy2.ipython
```

```python
%%R
library(ggplot2)
library(xkcd)
```

```python
%%R -w 500 -h 350 -u px
x <- seq(0,9.9,by=0.1)
y1 <- sin(x)
y2 <- y1*-1

dfr <- data.frame(color="Red", mood=x, time=y1)
dfd <- data.frame(color="Blue", mood=x, time=y2)

df <- rbind( dfr, dfd)

xrange <- range(x)
yrange <- range(y1)

p <- ggplot() + geom_line(data=df,aes(x=mood, y=time, color=color)) 
p + xkcdaxis(xrange,yrange) + 
     ylab("Y") +
     xlab("X")+
     ggtitle("ggplot2 xkcd") +
     theme(axis.ticks = element_blank(), axis.text.x = element_blank(), axis.text.y = element_blank())
```
